### Các đặc trưng thời gian

```python
#Chuyển sang datetime
data['ts'] = pd.to_datetime(data['ts'])

#Chuyển mili sang phút
data['min_played'] = round(data['ms_played'] / 60000, 2)

#Tách thời gian
data['date'] = data['ts'].dt.date
data['month'] = data['ts'].dt.month
data['year'] = data['ts'].dt.year
data['weekday'] = data['ts'].dt.day_name()
data['hour'] = data['ts'].dt.hour
data['minute'] = data['ts'].dt.minute

#Xoá cột vì đã tách ra
data.drop(columns=['ts', 'ms_played'], axis=1, inplace=True)

#Đặt date làm index
data.set_index('date', inplace=True)
```

### Top 10 hát hát, ca sĩ, album được nghe nhiều nhất

```python
top_track_name = data.groupby('track_name')['min_played'].sum().sort_values(ascending=False).head(10)
top_artist = data.groupby('artist_name')['min_played'].sum().sort_values(ascending=False).head(10)
top_album = data.groupby('album_name')['min_played'].sum().sort_values(ascending=False).head(10)
```

```python
import matplotlib.pylab as plt
fig, axes = plt.subplots(1, 2, figsize=(20,4))

# Đồ thị Top Artist
sns.barplot(
    x=top_artist.values,
    y=top_artist.index,
    ax=axes[0],
    hue=top_track_name.index, 
    palette='Greens_d',
    legend=False
)

axes[0].set_title('Top 10 Nhạc Sĩ Được Nghe Nhiều Nhất', fontsize=14)
axes[0].set_xlabel('Tổng thời gian nghe (phút)')
axes[0].set_ylabel('Tên nhạc sĩ')

#Top Album
sns.barplot(
    x=top_album.values,
    y=top_album.index,
    ax=axes[1],
    hue=top_track_name.index, 
    palette='Greens_d',
    legend=False
)
axes[1].set_title('Top 10 Album được nghe nhiều nhất', fontsize=14)
axes[1].set_xlabel('Tổng thời gian nghe (phút)')
axes[1].set_ylabel('Tên album')
plt.tight_layout()

#Top Track Name
plt.figure(figsize=(20, 4))
sns.barplot(
    x=top_track_name.values,
    y=top_track_name.index,
    hue=top_track_name.index, 
    palette='Greens_d',
    legend=False
)
plt.suptitle('Top 10 Bài Hát Được Nghe Nhiều Nhất', fontsize=14)
plt.xlabel('Tổng thời gian nghe (phút)')
plt.ylabel('Tên nhạc sĩ')
plt.tight_layout()
```

### Lượt nghe theo các đặc trưng thời gian (ngày, giờ) như thế nào ?

```python
data.head()
```

```python
hourly_play = data.groupby('hour')['min_played'].sum()
week_play = data.groupby('weekday')['min_played'].sum()
week_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
monthly_trend = data.groupby(['year', 'month'])['min_played'].sum().reset_index()
monthly_trend['year_month'] = pd.to_datetime(monthly_trend[['year', 'month']].assign(day=1))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(20, 5))

sns.lineplot(x=hourly_play.index, y=hourly_play.values,marker='o',ax=axes[0])
axes[0].set_title('Listening Behavior by Hour Of Day')
axes[0].set_ylabel('Total Minute')
axes[0].set_xticks(range(0, 24))
axes[0].grid(True)

sns.barplot(x=week_order, y=week_play[week_order],ax=axes[1])
axes[1].set_title('Listening Behavior by Day in Week')
axes[1].set_ylabel('Total Minutes')
axes[1].set_xlabel('Day in Week')
plt.tight_layout()

plt.figure(figsize=(20, 5))
sns.lineplot(data=monthly_trend, x='year_month', y='min_played', marker='o')
plt.title('Monthly Listening Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Total Minutes Played')
plt.grid(True)
plt.show()
```

```python
import statsmodels.api as sm
```

```python
decompostion = sm.tsa.seasonal_decompose(monthly_trend['min_played'].iloc[:90], model='additive', period=30)

plt.figure(figsize=(10, 7))

trend = decompostion.trend
seasonal = decompostion.seasonal
residual = decompostion.resid
oberved = decompostion.observed

plt.subplot(4, 1, 1)
plt.plot(oberved, label='Observed')
plt.legend()

plt.subplot(4, 1, 2)
plt.plot(trend, label='Trend')
plt.legend()

plt.subplot(4, 1, 3)
plt.plot(residual, label='Residual')
plt.legend()

plt.subplot(4, 1, 4)
plt.plot(seasonal, label='Seasonal')
plt.legend()

plt.tight_layout()
```
